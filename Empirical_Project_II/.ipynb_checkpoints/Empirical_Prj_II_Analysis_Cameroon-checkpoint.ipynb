## Question

Factors that influence prevalence of youth smoking in Cameroon:
+ Age
+ Sex
+ Availability of spending money
+ Exposure to smokers
    + Home
    + Indoor events outside home
    + Outdoor events

```python
%%stata
logistic smoker i.ageCategory i.female i.hasMoney i.easyToQuit i.smokingHome i.smokingIndoors i.smokingOutdoors
```

```python
%%stata
probit smoker i.ageCategory i.female i.hasMoney i.easyToQuit i.smokingHome i.smokingIndoors i.smokingOutdoors
```

```python
%%stata
margins i.ageCategory i.female i.hasMoney i.easyToQuit i.smokingHome i.smokingIndoors i.smokingOutdoors, atmeans
```
